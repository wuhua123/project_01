```python
# Combine Data by Market Cap

df_crypto_marketCap = pd.DataFrame({"Date": df_btc["Date"],
                          "BTC Cap":df_btc_2y["Market Cap"], 
                          "ETH Cap": df_eth_2y["Market Cap"],
                          "XRP Cap": df_xrp_2y['Market Cap'],
                          "BCH Cap": df_bch_2y['Market Cap'],
                          "EOS Cap": df_eos_2y['Market Cap'],
                          "XLM Cap": df_xlm_2y['Market Cap'],
                          "LTC Cap": df_ltc_2y['Market Cap'],
                          "ADA Cap": df_ada_2y['Market Cap'],
                          "XMR Cap": df_xmr_2y['Market Cap'],
                          "TRX Cap": df_trx_2y['Market Cap'],
                         })
# Combine data by Volume
df_crypto_volume = pd.DataFrame({"Date": df_btc_2y["Date"],
                          "BTC Vol": df_btc_2y["Volume"], 
                          "ETH Vol": df_eth_2y["Volume"],
                          "XRP Vol": df_xrp_2y['Volume'],
                          "BCH Vol": df_bch_2y['Volume'],
                          "EOS Vol": df_eos_2y['Volume'],
                          "XLM Vol": df_xlm_2y['Volume'],
                          "LTC Vol": df_ltc_2y['Volume'],
                          "ADA Vol": df_ada_2y['Volume'],
                          "XMR Vol": df_xmr_2y['Volume'],
                          "TRX Vol": df_trx_2y['Volume'],
                         })

df_crypto_volume.head()
```

```python
#Look for top 5 cryptos for the time being

# sort based on market cap
most_update = df_crypto_marketCap.loc[0,:]
most_update = most_update.drop("Date")
most_update_sort = most_update.sort_values(ascending=False )

# Finding the current top 5 crypto currencies based on Market Capital
top_five = most_update_sort.head()

top5_cryptos = pd.DataFrame({"Name": [top_five.index[i] for i in range(0,5)],
             "Market Capital":[top_five.values[i] for i in range(0,5)] })

for i in range(0, 5):
    top5_cryptos.loc[i, 'Name'] = str(top5_cryptos.loc[i,'Name'])
    short_name = top5_cryptos.loc[i, 'Name'].split()[0]
    top5_cryptos.loc[i, 'Name'] = short_name
    #top5_cryptos.to_string(top5_cryptos.loc[i, 'Name'])

print(top5_cryptos)
print(top5_cryptos['Name'].dtype)
```

```python
# Graph top 5 cryptos
%matplotlib notebook

#top5_cryptos_graph = top5_cryptos.set_index('Name')
x_axis = np.arange(len(top5_cryptos['Market Capital']))

plt.bar(x_axis, top5_cryptos['Market Capital'], color='blue', align='center')

tick_locations = [i for i in x_axis]
tick_names = top5_cryptos['Name']
plt.xticks(tick_locations, tick_names)

plt.title("Top 5 Cryptos for the time being")
plt.ylabel("Market Capital")

plt.savefig("graphs/Top5.png")
```

```python
#prepare for further data retrieving
#crypto_code = ['BTC', 'ETH', 'XRP', 'EOS', 'BCH', 'XLM', 'LTC', 'ADA', 'XMR', 'TRX']

all_cryptos_allPeriods ={'BTC': [df_btc, df_btc_2y],
                         'ETH': [df_eth, df_eth_2y], 
                         'XRP': [df_xrp, df_xrp_2y],
                         'EOS': [df_eos, df_eos_2y],
                         'BCH': [df_bch, df_bch_2y],
                         'XLM': [df_xlm, df_xlm_2y],
                         'LTC': [df_ltc, df_ltc_2y],
                         'ADA': [df_ada, df_ada_2y],
                         'XMR': [df_xmr, df_xmr_2y],
                         'TRX': [df_trx, df_trx_2y]}

#print(all_cryptos_allPeriods['BTC'])
cryptos_2y = []
cryptos_2y_names = []

for name in top5_cryptos['Name']:
    cryptos_2y.append(all_cryptos_allPeriods[name][1])
    cryptos_2y_names.append(name)

# calculate Daily Change Percentage for top 5 cryptos
for i in range(0,5):
    cryptos_2y[i]["Daily Change Percent(%)"] = ((cryptos_2y[i]["Close"] - cryptos_2y[i]["Close"].shift(-1))/cryptos_2y[i]["Close"].shift(-1))*100
    cryptos_2y[i] = cryptos_2y[i].dropna()
```

```python
#numpy.std- Standard Deviation
# Calculate 2 years Deviation for top 5 cryptos

cryptos_2ydev = []
for i in range(0,5):
    cryptos_2ydev.append(np.std(cryptos_2y[i]["Daily Change Percent(%)"], axis=0))
    print(cryptos_2y_names[i])
    print(cryptos_2ydev[i])
    print()
```

```python
%matplotlib notebook

cryptos_daily_c_pct = []

for i in range(0, 5):
    temp_df = cryptos_2y[i][['Date', 'Daily Change Percent(%)']]
    temp_df = temp_df.rename(columns={'Daily Change Percent(%)': cryptos_2y_names[i]})
    cryptos_daily_c_pct.append(temp_df)

temp_1 = pd.merge(cryptos_daily_c_pct[0], cryptos_daily_c_pct[1], on='Date')
temp_1 = temp_1.rename(columns={'Daily Change Percent(%)_x': cryptos_2y_names[0],
                                                  'Daily Change Percent(%)_y': cryptos_2y_names[1],})
temp_2 = pd.merge(cryptos_daily_c_pct[2], cryptos_daily_c_pct[3], on='Date')
temp_2 = temp_2.rename(columns={'Daily Change Percent(%)_x': cryptos_2y_names[2],
                                                  'Daily Change Percent(%)_y': cryptos_2y_names[3],})
temp_3 = pd.merge(temp_1, temp_2)
daily_c_pct_df = pd.merge(temp_3, cryptos_daily_c_pct[4], on = 'Date')

#print(daily_c_pct_df.head())

n_bins = 10
x = daily_c_pct_df[[cryptos_2y_names[0],cryptos_2y_names[1],cryptos_2y_names[2],\
                    cryptos_2y_names[3],cryptos_2y_names[4]]].as_matrix()
#print(x)

fig, ax = plt.subplots(nrows=1, ncols=1)
#ax0,ax1,ax2,ax3,ax4,ax5 = axes.flatten()

colors = ['red', 'blue', 'green', 'orange', 'purple']
labels = [cryptos_2y_names[0],cryptos_2y_names[1],cryptos_2y_names[2],cryptos_2y_names[3],cryptos_2y_names[4]]
ax.hist(x, n_bins, histtype='bar', color=colors, label=labels)
ax.legend(prop={'size': 10})
ax.set_title('frequency of top 5 cryptos in last 2 years')
```

```python
# resample daily change percentage weekly

plt.figure(figsize=(10,7)) 

colors = ['blue', 'red', 'orange', 'purple', 'green']
markers = ['s', 'o', 's', '^', 'x']

for i in range(0, 5):
    
    start_date = cryptos_2y[i]['Date'].min()
    start_date = start_date.strftime('%Y-%m-%d')
    end_date = cryptos_2y[i]['Date'].max()
    end_date = end_date.strftime('%Y-%m-%d')
    
    Index = pd.DatetimeIndex(start=start_date, end=end_date, freq="D")
    daily_change_pct = cryptos_2y[i]["Daily Change Percent(%)"].tolist()
    #s_cryptos_pct[i] = pd.DataFrame({"Daily Change Percent(%)": daily_change_pct}, index=Index)
    temp1 = pd.DataFrame({"Daily Change Percent(%)": daily_change_pct}, index=Index)
    
    #resample
    #cryptos_pct[i] = df_for_st.resample(rule="M",how=np.std)
    temp2 = temp1.resample(rule="M",how=np.std)
    #cryptos_pct_dev[i] = cryptos_pct[i]["Daily Change Percent(%)"]
    monthly_dev = temp2["Daily Change Percent(%)"]
    
    
    monthly_dev.plot(marker=markers[i], color=colors[i], label=cryptos_2y_names[i])
    plt.savefig("graphs/deviation/eos_PCT_dev.png")
  
    
plt.legend()

plt.title("Top 5 Crypto Deviation Analyzation")
plt.xlabel("Year and Months")
plt.ylabel("Monthly Deviation")
```
